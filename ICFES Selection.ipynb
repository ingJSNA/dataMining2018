```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# Read and combine files
p20171 = pd.read_csv("./data/ICFES/SB11-20171-RGSTRO-CLFCCN-V1-0.zip", sep="¬", engine="python")

p20172 = pd.read_csv("./data/ICFES/SB11-20172-RGSTRO-CLFCCN-V1-0.zip", sep="¬", engine="python")
```

```python
# Add missing columns
p20171['COLE_JORNADA'] = None

p20172['FAMI_TIENEINTERNET'] = None

#icfes.drop(columns=['ESTU_ESTUDIANTE', 'ESTU_COD_RESIDE_DEPTO', 'ESTU_COD_RESIDE_MCPIO', 'COLE_CODIGO_ICFES'])

# Rename columns
p20171 = p20171.rename(index=str, columns={"FAMI_COME_CARNEPESCADOHUEVO": "FAMI_COMECARNEPESCADOHUEVO",
                                           "FAMI_COME_CEREALFRUTOSLEGUMBRE": "FAMI_COMECEREALFRUTOSLEGUMBRE",
                                           "FAMI_TIENE_CONSOLAVIDEOJUEGOS": "FAMI_TIENECONSOLAVIDEOJUEGOS",
                                           "FAMI_COME_LECHEDERIVADOS": "FAMI_COMELECHEDERIVADOS",
                                          "FAMI_SITUACION_ECONOMICA": "FAMI_SITUACIONECONOMICA", 
                                          "FAMI_TIENE_SERVICIOTV": "FAMI_TIENESERVICIOTV",
                                          "FAMI_TRABAJOLABOR_MADRE": "FAMI_TRABAJOLABORMADRE", 
                                          "FAMI_TRABAJOLABOR_PADRE": "FAMI_TRABAJOLABORPADRE"})

p20172 = p20172.rename(index=str, columns={"x": "x", 
                                           "x": "x"})
```

```python
# Mergesort is the only stable algorithm
p1 = np.sort(p20171.columns.values, kind='mergesort')
p2 = np.sort(p20172.columns.values, kind='mergesort')

columns = pd.DataFrame(data=p1, columns={'p1'}).join(pd.DataFrame(data=p2, columns={'p2'}))

#display(columns[64: ])

# Find not matching columns
columns[columns['p1'] != columns['p2']]
```

```python
# Concat the two periods
icfes = pd.concat([p20171, p20172], ignore_index=True, join='outer')

icfes.describe()
```

```python
(icfes.isnull().sum() / len(icfes)).sort_values(ascending=False)
```

```python
icfes['FAMI_TRABAJOLABORPADRE'].describe(include='all')
```

```python
icfes['FAMI_TRABAJOLABORPADRE'].value_counts(dropna=False)
```

```python
# Replace extrange characters in string values
icfes.replace(to_replace=('\"', '`', '\''), value=('', ' ', ' '), regex=True, inplace=True)


#icfes[icfes['COLE_NOMBRE_ESTABLECIMIENTO'].str.contains("IDEBIC")].iloc[0:5, 14:]
```
