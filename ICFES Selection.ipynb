```python
import pandas as pd
import numpy as np
```

```python
# Read and combine files
p20171 = pd.read_csv("./data/ICFES/SB11-20171-RGSTRO-CLFCCN-V1-0.zip", sep="¬", engine="python")

p20172 = pd.read_csv("./data/ICFES/SB11-20172-RGSTRO-CLFCCN-V1-0.zip", sep="¬", engine="python")
```

```python
# Mergesort is the only stable algorithm
d = {'p1': np.sort(p20171.columns.values, kind='mergesort'), 'p2': np.sort(p20172.columns.values, kind='mergesort')}

columns = pd.DataFrame(data=d)

columns

# Find not matching columns
columns[columns['p1'] != columns['p2']]
```

```python
# Add missing columns

p20171['COLE_JORNADA'] = None

#icfes.drop(columns=['ESTU_ESTUDIANTE', 'ESTU_COD_RESIDE_DEPTO', 'ESTU_COD_RESIDE_MCPIO', 'COLE_CODIGO_ICFES'])

# Rename columns
p20171 = p20171.rename(index=str, columns={"A": "a", "B": "c"})

#p20172 = p20172.rename(index=str, columns={"ESTU_PAIS_RESIDE": "ESTU_PAIS_RESIDE", "B": "c"})
```

```python
# Concat the two periods
icfes = pd.concat([p20171, p20172], ignore_index=True)

icfes.head()
```

```python
# Replace extrange characters in string values
icfes.replace(to_replace=('\"', '`', '\''), value=('', ' ', ' '), regex=True, inplace=True)


#icfes[icfes['COLE_NOMBRE_ESTABLECIMIENTO'].str.contains("IDEBIC")].iloc[0:5, 14:]
```
